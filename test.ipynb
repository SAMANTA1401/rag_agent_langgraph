```python
from manim import *

class ParticleMotion(Scene):
    def construct(self):
        # Define the functions
        def a(t):
            return 0.3 * t + 2.4

        def v(t):
            return 0.15 * t**2 + 2.4 * t - 12

        def x(t):
            return 0.05 * t**3 + 1.2 * t**2 - 12 * t + 4.6

        # Time when particle is at rest
        t_rest = 4

        # Axes
        axes = Axes(
            x_range=(-5, 25, 1),
            y_range=(-25, 10, 1),
            x_axis_config={"numbers_to_include": [-5, 0, 4, 25]},
            y_axis_config={"numbers_to_include": [-25, 0, 4.6, 10]},
        )
        axes.add_coordinate_labels()
        self.play(Create(axes))

        # Graphs
        a_graph = axes.plot(a, x_range=(-5, 25), color=RED)
        v_graph = axes.plot(v, x_range=(-5, 25), color=GREEN)
        x_graph = axes.plot(x, x_range=(-5, 25), color=BLUE)

        a_label = axes.get_graph_label(a_graph, "a(t)", x_val=20)
        v_label = axes.get_graph_label(v_graph, "v(t)", x_val=20)
        x_label = axes.get_graph_label(x_graph, "x(t)", x_val=20)

        self.play(
            Create(a_graph),
            Create(v_graph),
            Create(x_graph),
            Create(a_label),
            Create(v_label),
            Create(x_label),
        )
        self.wait(2)

        # Particle animation
        particle = Dot(axes.coords_to_point(4.6, 0), color=YELLOW)
        particle_label = MathTex("Particle").next_to(particle, UP)
        self.play(Create(particle), Create(particle_label))

        # Follow the position
        def update_particle(obj):
            current_t = self.time  # Manim's time
            if current_t < 25:
                new_pos = axes.coords_to_point(x(current_t), 0)
                obj.move_to(new_pos)

        particle.add_updater(update_particle)
        self.play(MoveAlongPath(particle,Line(start=axes.coords_to_point(4.6, 0),end=axes.coords_to_point(x(25), 0))), run_time=25, rate_func=linear)
        particle.remove_updater(update_particle) # Stop updating

        # Mark the point where the particle stops
        rest_point = Dot(axes.coords_to_point(x(t_rest), 0), color=PURPLE)
        rest_label = MathTex(f"Rest at t={t_rest}s, x={x(t_rest):.2f}m").next_to(rest_point, DOWN)
        self.play(Create(rest_point), Create(rest_label))

        self.wait(5)
```

```python
import pygame
import pymunk
import numpy as np

# Initialize pygame
pygame.init()

# Set up the screen
WIDTH, HEIGHT = 800, 400
screen = pygame.display.set_mode((WIDTH, HEIGHT))
clock = pygame.time.Clock()

# Physics space
space = pymunk.Space()
space.gravity = (0, 0)  # No gravity since it's 1D motion

# Particle properties
mass = 1
radius = 10
x_start = 100  # Initial x position on screen
y_position = HEIGHT // 2  # Keep it centered

# Create the particle
body = pymunk.Body(mass, pymunk.moment_for_circle(mass, 0, radius))
body.position = (x_start, y_position)
shape = pymunk.Circle(body, radius)
shape.elasticity = 0.8
space.add(body, shape)

# Time parameters
t = 0
dt = 0.1  # Simulation time step
t_max = 5  # Max simulation time
x_values = []  # Store positions for visualization

# Acceleration function
def acceleration(t):
    return 0.3 * t + 2.4

# Velocity and position functions
v = -12.0
x = 4.60
while t < t_max:
    v += acceleration(t) * dt
    x += v * dt
    x_values.append((x, t))
    t += dt

# Main loop
running = True
while running:
    screen.fill((255, 255, 255))

    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    # Move particle along calculated positions
    if x_values:
        new_x, _ = x_values.pop(0)
        body.position = (new_x * 20 + WIDTH // 2, y_position)  # Scale for visibility

    # Draw particle
    pygame.draw.circle(screen, (255, 0, 0), (int(body.position.x), int(body.position.y)), radius)

    pygame.display.flip()
    clock.tick(60)
    space.step(dt)

pygame.quit()
```

```python
import pygame
import pymunk
import pymunk.pygame_util
import math

# Initialize Pygame
pygame.init()

# Screen settings (Larger Canvas)
WIDTH, HEIGHT = 800, 400  # Increased size for better visualization
screen = pygame.display.set_mode((WIDTH, HEIGHT))
clock = pygame.time.Clock()
font = pygame.font.SysFont(None, 30)

# Pymunk Space
space = pymunk.Space()
space.gravity = (0, -300)  # Gravity (increased to match new scale)

# Convert physics coordinates to Pygame coordinates
def to_pygame(p):
    return int(p[0]), int(HEIGHT - p[1])

# Projectile parameters
initial_x, initial_y = 50, 50
angle = math.radians(45)  # Launch angle in radians
velocity = 400  # Increased speed for larger range and max height
vx = velocity * math.cos(angle)
vy = velocity * math.sin(angle)

# Create projectile
body = pymunk.Body(1, pymunk.moment_for_circle(1, 0, 10))
body.position = initial_x, initial_y
body.velocity = vx, vy
shape = pymunk.Circle(body, 10)
shape.elasticity = 0.5
space.add(body, shape)

# Store trajectory points
trajectory = []
max_height = initial_y
final_range = None  # Track the final range when projectile lands

# Draw velocity vectors


# Draw axes
def draw_axes():
    pygame.draw.line(screen, (0, 0, 0), (50, HEIGHT - 50), (WIDTH - 50, HEIGHT - 50), 3)  # X-axis (Range)
    pygame.draw.line(screen, (0, 0, 0), (50, HEIGHT - 50), (50, 50), 3)  # Y-axis (Height)

    # Labels
    x_label = font.render("Range (X)", True, (0, 0, 0))
    y_label = font.render("Height (Y)", True, (0, 0, 0))
    screen.blit(x_label, (WIDTH - 150, HEIGHT - 40))
    screen.blit(y_label, (60, 20))

# Draw max height and range markers
def draw_markers():
    global max_height, final_range
    px, py = to_pygame(body.position)

    # Track max height
    if body.velocity.y < 0 and body.position.y > max_height:
        max_height = body.position.y

    # Max height marker
    max_h_y = to_pygame((0, max_height))[1]
    pygame.draw.line(screen, (0, 0, 255), (50, max_h_y), (WIDTH - 50, max_h_y), 2)
    height_label = font.render("Max Height", True, (0, 0, 255))
    screen.blit(height_label, (WIDTH - 200, max_h_y - 20))

    # Mark final range when projectile lands
    if final_range is not None:
        pygame.draw.line(screen, (255, 0, 255), (final_range, HEIGHT - 50), (final_range, HEIGHT - 40), 3)
        range_label = font.render("Range", True, (255, 0, 255))
        screen.blit(range_label, (final_range - 30, HEIGHT - 80))

# Draw trajectory trace
def draw_trajectory():
    if len(trajectory) > 1:
        pygame.draw.lines(screen, (0, 0, 255), False, trajectory, 2)
        
        
def draw_vectors():
    px, py = to_pygame(body.position)
    vx, vy = body.velocity

    scale_factor = 0.15  # Adjusted for better visualization

    # Draw velocity components
    pygame.draw.line(screen, (255, 0, 0), (px, py), (px + vx * scale_factor, py), 3)  # vx (horizontal)
    pygame.draw.line(screen, (0, 255, 0), (px, py), (px, py - vy * scale_factor), 3)  # vy (vertical)

    # Labels for velocity components
    vx_label = font.render(f"vx", True, (255, 0, 0))
    vy_label = font.render(f"vy", True, (0, 255, 0))
    screen.blit(vx_label, (px + 20, py))
    screen.blit(vy_label, (px - 30, py - 30))

    #  Draw launch angle trace
    angle_length = 50  # Fixed length for better visualization
    angle_x = initial_x + angle_length * math.cos(angle)
    angle_y = initial_y + angle_length * math.sin(angle)

    pygame.draw.line(screen, (255, 165, 0), to_pygame((initial_x, initial_y)), to_pygame((angle_x, angle_y)), 3)
    
    # Label for angle trace
    angle_label = font.render(f" = {math.degrees(angle):.1f}", True, (255, 165, 0))
    screen.blit(angle_label, to_pygame((initial_x + 40, initial_y + 40)))


# Main loop
running = True
while running:
    screen.fill((255, 255, 255))
    
    draw_axes()
    draw_markers()
    draw_vectors()
    draw_trajectory()

    # Save trajectory points
    trajectory.append(to_pygame(body.position))

    # Step simulation
    space.step(1 / 110)

    # Draw projectile
    pygame.draw.circle(screen, (0, 0, 255), to_pygame(body.position), 10)

    # Check if projectile has landed
    if body.position.y <= 100 and body.velocity.y < 0 and final_range is None:
        final_range = to_pygame(body.position)[0]

    pygame.display.flip()
    clock.tick(60)

    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

pygame.quit()
```
